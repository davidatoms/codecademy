```python
# Create the Total Goals column
WorldCupMatches_df["Total Goals"] = WorldCupMatches_df["Home Team Goals"] + WorldCupMatches_df["Away Team Goals"]

sns.set_style("whitegrid")
sns.set_context("poster", font_scale=0.8)
f, ax = plt.subplots(figsize=(12, 7))
ax = sns.barplot(x=WorldCupMatches_df["Year"], y=WorldCupMatches_df["Total Goals"])
ax.set_title("Average Goals Per Year in World Cups")
plt.show()
```
